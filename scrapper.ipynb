```python
q=1
try:
    import requests
    from bs4 import BeautifulSoup
    from csv import writer
except Exception as e:
    print("Some modules are Missing {}".format(e))
#url = 'http://www.laughfactory.com/jokes/blonde-jokes/'+str(q)
#url = 'http://www.laughfactory.com/jokes/animal-jokes/'+str(q)
#url = 'http://www.laughfactory.com/jokes/clean-jokes/'+str(q)
#url = 'http://www.laughfactory.com/jokes/national-jokes/'+str(q)
#url = 'http://www.laughfactory.com/jokes/office-jokes/'+str(q)
#url = 'http://www.laughfactory.com/jokes/pop-culture-jokes/'+str(q)
url = 'https://www.rd.com/jokes/corny/'+str(q)
#url = 'http://www.laughfactory.com/jokes/political-jokes/'+str(q)
#url = 'http://www.laughfactory.com/jokes/miscellaneous-jokes/'+str(q)
#url = 'http://www.laughfactory.com/jokes/insult-jokes/'+str(q)
#url = 'http://www.laughfactory.com/jokes/how-to-be-insulting/'+str(q)
#url = 'http://www.laughfactory.com/jokes/holiday-jokes/'+str(q)
#url = 'http://www.laughfactory.com/jokes/food-jokes/'+str(q)
#url = 'http://www.laughfactory.com/jokes/family-jokes/'+str(q)
#url = 'http://www.laughfactory.com/jokes/boycott-these-jokes/'+str(q)
headers = {
'Accept-Encoding': 'gzip, deflate, sdch',
'Accept-Language': 'en-US,en;q=0.8',
'User-Agent': 'Mozilla/5.0 (Macintosh; Intel Mac OS X 10_10_1) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/39.0.2171.95 Safari/537.36',
'Accept': 'text/html,application/xhtml+xml,application/xml;q=0.9,image/webp,*/*;q=0.8',
'Referer': 'http://www.wikipedia.org/',
'Connection': 'keep-alive',}


r =requests.get(url, headers)
##print(r.text)
data = []
final=[]

soup = BeautifulSoup(r.text, 'html.parser')
#print(soup)
div = soup.find('div',class_="jokes-sec")
for x in div.findAll('div', class_="joke-text"):
    data.append(x.text.strip())

for i ,x in enumerate(data):
    #print(i,x)
    final=[x]
    #print(final)
    #break

    with open('newevent.csv', 'a') as f_object:
        writer_object = writer(f_object)

    # Pass the list as an argument into
    # the writerow()
        writer_object.writerow(final)

    #Close the file object
        f_object.close()  

# print(data)





```

```python
import pandas as pd
import glob
import os

# setting the path for joining multiple files
files = os.path.join("./", "data*.csv")

# list of merged files returned
files = glob.glob(files)

print("Resultant CSV after joining all CSV files at a particular location...");

# joining files with concat and read_csv
df = pd.concat(map(pd.read_csv, files), ignore_index=True)
print(df)
```
